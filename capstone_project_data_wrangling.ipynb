```python
import pandas as pd
```

```python
# url link to the csv file for importing college scorecard data
url = 'https://ed-public-download.app.cloud.gov/downloads/Most-Recent-Cohorts-All-Data-Elements.csv'
```

```python
# reading data into a dataframe 
data = pd.read_csv(url)
```

```python
# checking head of the data
data.head()
```

```python
# visually inspecting the data
data.info()
```

```python
data.describe()
```

```python
# Several values in the columns are represented as PrivacySuppressed: replacing them with NULL
data = data.replace('PrivacySuppressed', 'NULL')
```

```python
#Calculating percentage of NULL data in each column and selecting those that have less than 5% NULL values
columns_2 = data.isnull().sum()/data.shape[0]*100 < 5
```
